```python
from sklearn.linear_model import LinearRegression
```

```python
#Split data
```

```python
train_hd = hd.sample(frac=0.8, random_state = 24)
test_hd = hd.drop(train_hd.index)
```

```python
train_hd
```

```python
test_hd
```

```python
hd_model = LinearRegression()   #create Linear model
```

```python
hd.columns
```

```python
hd_model.fit(train_hd[['Apparent Temperature (C)']], train_hd['Humidity'])       #train training data model
```

```python
hd_pred = hd_model.predict(test_hd[['Apparent Temperature (C)']])  #make predictions on test data,i.e., predicted humidity values
```

```python
hd_pred
```

```python
actual_hd = test_hd['Humidity']    #actual humidity values 
```

```python
actual_hd
```

```python
from sklearn.metrics import mean_squared_error           #MSE is the avg squared difference between the predicted and actual humidity values.

mse = mean_squared_error(test_hd['Humidity'], hd_pred)      #Lower MSE = better accuracy since 0 means "the best"
mse
```

```python
hd_model.fit(train_hd[['Pressure (millibars)']], train_hd['Humidity'])
hd_pressure_pred = hd_model.predict(test_hd[['Pressure (millibars)']])

mean_squared_error(test_hd['Humidity'], hd_pressure_pred)
```
